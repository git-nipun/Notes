# Recursion

Recursion : A function calling itself.
n! = n * n-1 * n-2 * n-3 ........
n! = n * (n-1)!
fact(n) = n * fact(n-1)
fact(n-1) = (n-1) * fact(n-2)

Recursion is when a function is calling itself with a smaller input, which basically is easier to calculate.
To a trivial case which is easy or known.

Like in factorial base case, 0! = 1! = 1



```python
# 1st Way
def fact(n):
    
    # Base Case
    if n == 0:
        return 1
    
    return n * fact(n-1)

n = int(input())
fact(n)
```

```python
# 2nd way
def fact_2(n):
    
    # Base Case
    if n == 0:
        return 1
    
    small_output = fact(n-1)
    return n * small_output

n = int(input())
fact(n)
```

```python
# Q. Power of a Number
def power(base,exp):
    if exp == 0:
        return 1
    return base * power(base,exp-1)

base,exp = [int(i) for i in input().split()]
print(power(base,exp))
```

Recursion is based on a  mathematical concept of PMI(Principal of mathematical induction)



```python
# print sum of N natural numbers
def sum_n(n):
    if n == 0:
        return 0
    small_output = sum_n(n-1)
    output = n + small_output
    return output

n = int(input())
print(sum_n(n))
```

```python
# print first N natural numbers

def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n-1)
    print(n)
    return
    
n = int(input())
print_1_to_n(n)
```

```python
# print first N to 1 

def print_n_to_1(n):
    if n == 0:
        return
    print(n)
    print_n_to_1(n-1)
    return
    
n = int(input())
print_n_to_1(n)
```

```python
# Fibonnaci Sum
def fib(n):
    if n == 1 or n == 2:
        return 1
    fib_n_1 = fib(n-1)
    fib_n_2 = fib(n-2)
    output = fib_n_1 + fib_n_2
    return output

n = 4
fib(n)
```

```python
class Mobile:
    def __init__(self):
        self.model = 'RealMe X'
    def show_model(self):
        print("Model:",self.model)

realme = Mobile() # object/Instance creation
realme.show_model()
print(realme.model)

#changing the value
realme.model = "Realme Pro2"
print(realme.model)
realme.show_model()
```

```python
class Mobile:
    def __init__(self,m):
        self.model = m
        
    def show_details(self,p):
        self.price = p
        print("Model:",self.model,"Price",self.price)

realme = Mobile('Realme X')
realme.show_details(10000)
```

## Types of Variable
- Instance variable
- Class variable/ Static variable

```python
#Instance variable
class Mobile:
    def __init__(self):
        self.model = "Realme X" # Instance Variable
        
    def show_model(self):
        print(self.model) # Accessing Instance Variable
realme = Mobile()
print(realme.model)

```

```python
# Class variable or Static variable

class Mobile:
    fp = 'yes'     # class variable
    
    @classmethod
    def is_fp(cls):
        print(cls.fp)
        
realme = Mobile()
print(Mobile.fp)
```

```python
class Mobile:
    fp = 'yes'     # class variable
    
    @classmethod
    def is_fp(cls):
        print(cls.fp)    # Accessing Class Variable
        
realme = Mobile()
# print(Mobile.fp)
Mobile.is_fp()
```

```python
# Class Variable or Static Variable with Class Method
class Mobile:
    fp = 'yes'     # class variable
    
    def __init__(self):
        self.model = 'Realme X'             # Instance Variable
        
    def show_model(self):                   # Instance Method
        print("Model: ",self.model)         # Accessing Instance Variable
    
    @classmethod
    def is_fp(cls):       # class method
        print("Finger Print: ",cls.fp)   # Accessing class Variable
        
realme = Mobile()
realme.show_model()
Mobile.is_fp()
print()
Mobile.fp = 'No'
Mobile.is_fp()
```
